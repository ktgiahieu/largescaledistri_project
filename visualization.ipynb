```python
import matplotlib.pyplot as plt
import pandas as pd
import asyncio
import seaborn as sns
%matplotlib widget
fig = plt.figure()

async def update():
    for i in range(1):
        
        plt.clf()
        
        df = pd.read_csv('output.csv')
        # change df game names to be more readable
        df['game'] = df['game'].replace('Counter-Strike%3A%20Global%20Offensive', 'CS:GO')
        df['game'] = df['game'].replace('Just%20Chatting', 'Just Chatting')
        sns.scatterplot(data=df, x='polarity', y='subjectivity', hue='prediction', style='game', s=50, palette=sns.color_palette("tab10", 5))
        
        plt.xlabel('Polarity')
        plt.ylabel('Subjectivity')
        plt.title('Sentiment Analysis of Twitch Chat')
        plt.legend(loc='center left', bbox_to_anchor=(0.82, 0.35))

        fig.canvas.draw()
        await asyncio.sleep(1)


loop = asyncio.get_event_loop()
loop.create_task(update())
```
